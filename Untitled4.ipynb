```python
import pandas as pd
import matplotlib.pyplot as plt
TotalSubcatProfits['ProfitRatio'] = TotalSubcatProfits['Profit'] / TotalSubcatProfits['Sales']
plt.figure('figure 1', figsize=(12, 8))  # Set the figure name here
sub_categories = TotalSubcatProfits.index
profits = TotalSubcatProfits['Profit']
sales = TotalSubcatProfits['Sales']
profit_ratios = TotalSubcatProfits['ProfitRatio']
bar_width = 0.25
index = range(len(sub_categories))
plt.bar(index, profits, bar_width, label='Total Profit', color='skyblue')
plt.bar([i + bar_width for i in index], sales, bar_width, label='Sales', color='orange')
plt.bar([i + 2*bar_width for i in index], profit_ratios, bar_width, label='Profit Ratio', color='lightgreen')
plt.title('Total Profits, Sales, and Profit Ratio by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Value')
plt.xticks([i + bar_width for i in index], sub_categories, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
